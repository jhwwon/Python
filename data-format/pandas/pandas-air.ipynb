```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. 데이터 생성 - 간단한 학생 성적 데이터
data = {
    '이름': ['김철수', '이영희', '박민수', '최지영', '정다은'],
    '수학': [85, 92, 78, 95, 88],
    '영어': [90, 87, 85, 93, 91],
    '과학': [88, 89, 92, 87, 94],
    '학년': [1, 2, 1, 3, 2]
}

# DataFrame 생성
df = pd.DataFrame(data)

print("=== 생성된 데이터 ===")
print(df)
print()

# 2. 기본 정보 확인
print("=== 데이터 기본 정보 ===")
print(f"데이터 크기: {df.shape}")
print(f"컬럼명: {list(df.columns)}")
print()

# 3. 통계 요약
print("=== 성적 통계 요약 ===")
print(df[['수학', '영어', '과학']].describe())
print()

# 4. 새로운 컬럼 생성 - 평균 점수
df['평균'] = (df['수학'] + df['영어'] + df['과학']) / 3
df['평균'] = df['평균'].round(1)  # 소수점 1자리까지

print("=== 평균 점수 추가 후 ===")
print(df)
print()

# 5. 데이터 필터링 - 평균 90점 이상 학생
high_score_students = df[df['평균'] >= 90]
print("=== 평균 90점 이상 학생 ===")
print(high_score_students[['이름', '평균']])
print()

# 6. 정렬 - 평균 점수 순으로
df_sorted = df.sort_values('평균', ascending=False)
print("=== 평균 점수 순 정렬 ===")
print(df_sorted[['이름', '평균']])
print()

# 7. 그룹별 분석 - 학년별 평균
grade_avg = df.groupby('학년')[['수학', '영어', '과학', '평균']].mean()
print("=== 학년별 평균 점수 ===")
print(grade_avg.round(1))
print()

# 8. 최고점, 최저점 찾기
print("=== 과목별 최고/최저점 ===")
subjects = ['수학', '영어', '과학']
for subject in subjects:
    max_student = df.loc[df[subject].idxmax()]
    min_student = df.loc[df[subject].idxmin()]
    print(f"{subject} - 최고: {max_student['이름']}({max_student[subject]}점), "
          f"최저: {min_student['이름']}({min_student[subject]}점)")
print()

# 9. 간단한 시각화
plt.figure(figsize=(10, 6))

# 학생별 평균 점수 막대 그래프
plt.subplot(1, 2, 1)
plt.bar(df['이름'], df['평균'], color=['skyblue', 'lightgreen', 'coral', 'gold', 'pink'])
plt.title('학생별 평균 점수')
plt.ylabel('점수')
plt.xticks(rotation=45)

# 과목별 전체 평균
subject_means = df[['수학', '영어', '과학']].mean()
plt.subplot(1, 2, 2)
plt.bar(subject_means.index, subject_means.values, color=['red', 'blue', 'green'])
plt.title('과목별 전체 평균')
plt.ylabel('평균 점수')

plt.tight_layout()
plt.show()

# 10. CSV 파일로 저장
df.to_csv('student_scores.csv', index=False, encoding='utf-8-sig')
print("데이터가 'student_scores.csv' 파일로 저장되었습니다!")

# 11. 추가 분석 - 성적 등급 부여
def get_grade(score):
    if score >= 95:
        return 'A+'
    elif score >= 90:
        return 'A'
    elif score >= 85:
        return 'B+'
    elif score >= 80:
        return 'B'
    else:
        return 'C'

df['등급'] = df['평균'].apply(get_grade)

print("=== 최종 결과 (등급 포함) ===")
print(df[['이름', '평균', '등급']])

# 등급별 인원수
print("\n=== 등급별 분포 ===")
print(df['등급'].value_counts().sort_index())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1.   -    
data = {
    '': ['', '', '', '', ''],
    '': [85, 92, 78, 95, 88],
    '': [90, 87, 85, 93, 91],
    '': [88, 89, 92, 87, 94],
    '': [1, 2, 1, 3, 2]
}

# DataFrame 
df = pd.DataFrame(data)

print("===   ===")
print(df)
print()

# 2.   
print("===    ===")
print(f" : {df.shape}")
print(f": {list(df.columns)}")
print()

# 3.  
print("===    ===")
print(df[['', '', '']].describe())
print()

# 4.    -  
df[''] = (df[''] + df[''] + df['']) / 3
df[''] = df[''].round(1)  #  1

print("===     ===")
print(df)
print()

# 5.   -  90  
high_score_students = df[df[''] >= 90]
print("===  90   ===")
print(high_score_students[['', '']])
print()

# 6.  -   
df_sorted = df.sort_values('', ascending=False)
print("===     ===")
print(df_sorted[['', '']])
print()

# 7.   -  
grade_avg = df.groupby('')[['', '', '', '']].mean()
print("===    ===")
print(grade_avg.round(1))
print()

# 8. ,  
print("===  / ===")
subjects = ['', '', '']
for subject in subjects:
    max_student = df.loc[df[subject].idxmax()]
    min_student = df.loc[df[subject].idxmin()]
    print(f"{subject} - : {max_student['']}({max_student[subject]}), "
          f": {min_student['']}({min_student[subject]})")
print()

# 9.  
plt.figure(figsize=(10, 6))

#     
plt.subplot(1, 2, 1)
plt.bar(df[''], df[''], color=['skyblue', 'lightgreen', 'coral', 'gold', 'pink'])
plt.title('  ')
plt.ylabel('')
plt.xticks(rotation=45)

#   
subject_means = df[['', '', '']].mean()
plt.subplot(1, 2, 2)
plt.bar(subject_means.index, subject_means.values, color=['red', 'blue', 'green'])
plt.title('  ')
plt.ylabel(' ')

plt.tight_layout()
plt.show()

# 10. CSV  
df.to_csv('student_scores.csv', index=False, encoding='utf-8-sig')
print(" 'student_scores.csv'  !")

# 11.   -   
def get_grade(score):
    if score >= 95:
        return 'A+'
    elif score >= 90:
        return 'A'
    elif score >= 85:
        return 'B+'
    elif score >= 80:
        return 'B'
    else:
        return 'C'

df[''] = df[''].apply(get_grade)

print("===   ( ) ===")
print(df[['', '', '']])

#  
print("\n===   ===")
print(df[''].value_counts().sort_index())
```

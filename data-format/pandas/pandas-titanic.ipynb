```python
# 나이가 35살 초과한 사람의 데이터를 또 다른 데이터프레임 변수에 저장
above_35 = titanic[titanic["Age"] > 35]
above_35
```

```python
# PClass가 2 혹은 3인 행을 또 다른 데이터 프레임으로 변경
class_23 = titanic[titanic["Pclass"].isin([2, 3])]
class_23 = titanic[(titanic["Pclass"] == 2) | (titanic["Pclass"] == 3)]
class_23
```

```python
titanic['Age'].info()
```

```python
# Age 시리즈에 데이터가 있는 행들만 가져오기
age_no_na = titanic[titanic["Age"].notna()]
age_no_na
```

```python
# Age 시리즈에 데이터가 없는 행들만 가져오기
age_no_na = titanic[titanic["Age"].isna()]
age_no_na
```

```python
# 35세 초과의 승객의 이름
adult_names = titanic.loc[titanic["Age"] > 35, "Name"]
adult_names.describe()
```
